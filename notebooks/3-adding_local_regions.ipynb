# Loading and inspecting data

```python
import pandas as pd
import numpy as np
```

```python
file_path1='./sla-preprocessing-data/merged_data_26_feb.xlsx' # investor data
file_path2='./sla-preprocessing-data/countries_regions_codes.xlsx' # country codes table
file_path3='./sla-preprocessing-data/regions_202202261659.csv' # local regions table
file_path4='./sla-preprocessing-data/investor_deals.xlsx' # target data
```

## merged_data

```python
merged_data=pd.read_excel(file_path1)
merged_data=merged_data.drop(['Unnamed: 0'],axis=1)
```

## country_codes

```python
country_codes=pd.read_excel(file_path2)
country_codes=country_codes.drop(['Unnamed: 0'], axis=1)
```

## loc_region_codes

```python
loc_region_codes=pd.read_csv(file_path3, encoding='utf-8', delimiter=';')
loc_region_codes=loc_region_codes.drop(['name_de','name_fr','name_es'],axis=1)
loc_region_codes.columns=['local_region_id', 'country_id', 'local_region_name']
```

## target_data

```python
target_data=pd.read_excel(file_path4)
```

```python
target_region=target_data[['deal_id', 'target_country_id', 'target_region_id']]
```

```python
target_region['target_country_id'] = target_region['target_country_id'].fillna(0).astype(int)
target_region['target_region_id'] = target_region['target_region_id'].fillna(0).astype(int)
```

# Matching location data

```python
target_matched = pd.merge(target_region, country_codes, left_on='target_country_id',right_on='country-code', how='left')
target_matched = pd.merge(target_matched, loc_region_codes, left_on='target_region_id', right_on='local_region_id', how='left')
target_matched = target_matched[['deal_id', 'target_country_id', 'target_region_id', 'region', 'country', 'local_region_name']]
target_matched.columns=['deal_id', 'target_country_id', 'target_region_id', 'target_region', 'target_country', 'target_local_region']
```

## OHE region

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse = False)

ohe.fit(target_matched[['target_region']])

target_encoded = ohe.transform(target_matched[['target_region']])

encoded_df = target_matched.copy()

encoded_df['target_Eastern_Europe'], encoded_df['target_Northern_Africa'], encoded_df['target_Northern_America'], encoded_df['target_Northern_Europe'], encoded_df['target_Southern_Europe'], encoded_df['target_Western_Europe'], encoded_df['NaN_region'] = target_encoded.T
```

## OHE country

```python
ohe = OneHotEncoder(sparse = False)

ohe.fit(target_matched[['target_country']])

target_country_encoded = ohe.transform(target_matched[['target_country']])

encoded_df['target_Austria'], encoded_df['target_Belgium'], encoded_df['target_Czechia'], encoded_df['target_Egypt'], encoded_df['target_Germany'], encoded_df['target_Italy'], encoded_df['target_Netherlands'],encoded_df['target_Norway'],encoded_df['target_Poland'],encoded_df['target_Portugal'],encoded_df['target_Romania'],encoded_df['target_Slovakia'],encoded_df['target_Spain'],encoded_df['target_Switzerland'],encoded_df['target_United_States_of_America'],encoded_df['NaN_country']= target_country_encoded.T
```

## OHE local region

```python
ohe = OneHotEncoder(sparse = False)

ohe.fit(target_matched[['target_local_region']])

target_state_encoded = ohe.transform(target_matched[['target_local_region']])

encoded_df['target_Baden-WÃ¼rttemberg'], encoded_df['target_Bavaria'], encoded_df['target_Berlin'], encoded_df['target_Brandenburg'], encoded_df['target_Bremen'], encoded_df['target_Hamburg'],encoded_df['target_Hesse'],encoded_df['target_Lower_Saxony'],encoded_df['target_Mecklenburg-Vorpommern'],encoded_df['target_North_Rhine-Westphalia'],encoded_df['target_Rhineland-Palatinate'],encoded_df['target_Saarland'],encoded_df['target_Saxony'],encoded_df['target_Saxony-Anhalt'],encoded_df['target_Schleswig-Holstein'],encoded_df['target_Thuringia'],encoded_df['NaN_state'] = target_state_encoded.T

encoded_df = encoded_df.drop(['target_country_id', 'target_region_id', 'target_region', 'target_country', 'target_local_region', 'NaN_country', 'NaN_region', 'NaN_state'], axis=1)

encoded_df = encoded_df.groupby('deal_id').max().reset_index()
encoded_df
```

# Merge all data and drop columns

```python
merged_data=merged_data.drop(['company_category_id','company_state_id','fees_rate','inroad','exit_oriented_y','management_takeover_y','seriousness','agnostic_y','margin','created_at','updated_at','company_inroad_id','company_id','max_revenue','min_revenue','max_ebitda','min_ebitda','max_equity','min_equity','searchmandate','invest_profile_id'], axis=1)

final_df = pd.merge(merged_data, encoded_df, on='deal_id', how='left')

final_df_no_names = final_df.drop(['investor_name', 'name'],axis=1)

final_df_no_names = final_df_no_names.fillna(0)
```
